### 6) [P] Generate a scatterplot again, however, now color the points based off of the labels generated by the clustering, and NOT the correct labels. Add the centroids using a distinguishing color and size with an alpha=0.5. Your plot should look something like the following:

````python
#Ref: https://stackoverflow.com/questions/54389331/adjust-different-transparency-for-different-class-in-seaborn-scatter-plot

# df_preds is dataframe containing coordinates of points and k-means prediction
# centroids is np array containing coordinates of centroids for k-means
def generate_scatter(df_preds, centroids):
    palette_list = ["b","orange","g","r","purple"]
    df_centroids = pd.DataFrame({"x1":centroids[:,0], "x2":centroids[:,1],"Centroid":centroids.shape[0]*[1]}) # use centroids.shape[0] to get # of clusters
    # Plot centroid scatter plot on top of predictions scatter plot
    ax = sns.scatterplot(x="x1", y="x2", hue="pred", data=df_preds, palette=palette_list[:centroids.shape[0]]) # use centroids.shape[0] to get # of clusters
    sns.scatterplot(x="x1", y="x2", data=df_centroids, hue="Centroid", size="Centroid", sizes=(250,250),palette=["grey"], alpha=0.5, legend=False, ax=ax)
    plt.title("Clustered Labels (k={})".format(centroids.shape[0]))
    plt.show()
````

````python
centroids = kmeans.cluster_centers_
df_data_pred = pd.DataFrame({"x1":df_data.x1, "x2":df_data.x2, "pred":kmeans_pred})
generate_scatter(df_data_pred, centroids)
````

### 8) [P] Let's take a step backward and assume that you are uncertain about how many actual clusters you have. Evaluate cluster sizes from 2 through 10. Create a data frame that stores the SSE and the silhouette score for each k. Show the resulting data frame.

````python
# clusters is an integer representing number of clusters for kmeans
def get_cluster_results(clusters):
    kmeans=KMeans(n_clusters=clusters)
    kmeans_fit = kmeans.fit(df_data.iloc[:,:2])
    kmeans_pred = kmeans.fit_predict(df_data.iloc[:,:2])
    return [kmeans.inertia_, silhouette_score(df_data.iloc[:,:2],kmeans_pred)]

cluster_results = np.array([get_cluster_results(i) for i in range(2,11)])   
df_cluster_results = pd.DataFrame({"k": list(range(2,11)), "SSE":cluster_results[:,0], "sil_score":cluster_results[:,1]})
df_cluster_results
````

### 11) [P] Now, re-generate your clustering for k=5. You will likely notice that your labels will be different colors than your original labels. That's annoying. Write the code to remap your cluster labels to use the same order as your labels as your ground truth. And, re-generate the correct plot, AND the clustered plot. This will make it easier to distinguish the observations that have different labels between the plots.
### (HINT: How? Perhaps the easiest approach is to use the contingency_matrix function in sklearn.metrics.cluster package. Show the output of that, then think about how you can use that to get the best mapping.)
### If you do it correctly, you should have two plots that look very, very similar: Your plots might be slightly different, but you should see at least a couple of points that were partitioned into the wrong cluster.

````python
from sklearn.metrics.cluster import contingency_matrix
# Find max index of each row and create dictionary to translate index of max column to index of row for each entry in preds
def remap_preds(preds):
    return_preds = len(preds)*[0]
    matrix = np.asarray(contingency_matrix(df_data.target, preds))
    max_array = np.argmax(matrix,axis=1)
    t_dict = dict(zip(max_array, list(range(len(max_array))))) # create translation dictionary to convert prediction to original label
    for i in range(len(preds)):
        return_preds[i] = t_dict[preds[i]]
    return return_preds
````

````python
sns.scatterplot(x="x1", y="x2", hue="target", data=df_data)
plt.title("Actual Labels of 5 Blobs")
plt.show()
````

````python
kmeans=KMeans(n_clusters=5)
kmeans_fit = kmeans.fit(df_data.iloc[:,:2])
kmeans_pred = kmeans.fit_predict(df_data.iloc[:,:2])
df_data_pred = pd.DataFrame({"x1":df_data.x1, "x2":df_data.x2, "pred":remap_preds(kmeans_pred)}) # call remap_preds to get true labels
generate_scatter(df_data_pred, kmeans.cluster_centers_)
````

### 13) [M] Since we do have ground truth, there are several metrics that are used to assess the quality of your clustering. Read the section on clustering performance evaluation in Scikit-learn's documentation page: https://scikit-learn.org/stable/modules/clustering.html#clustering-evaluation
### Pay close attention to the following: Adjusted Rand Index, Homogeneity, Completeness, and V-measure. In this section, very briefly describe each of these four measures. Though we did not explicitly cover them in class, these are good to know.

**Adjusted Rand Index** - measures the similarity of true and predicted labels while ignoring permutation and with chance normalization. It is also symmetric, so swapping the argument does not change the score. Ranges from [-1,1], where 1 is perfect labeling. Advantages are that no assumption is made on the cluster structure and random labeling has a score close to 0. An obvious disadvantage is that ARI requires the knowledge of the ground truth classes.   
**Homogeneity** - each cluster contains only members of a single class [0,1]  
**Completeness** - all members of a given class are assigned to the same cluster [0,1]  
**V-measure** - The harmonic mean of homogeneity and completeness [0,1]:  
            v = ((1+B) * homogeneity * completeness)/(B * homogeneity + completeness)  
            B > 1 more weight is attributed to completeness  
            B < 1 more weight is attributed to homogeinity


### 14) [P] Create a new data frame that contains these four measures for all KMeans clusterings of k between 2-10. As before, show your data frame, then generate a plot for each. However, since each of these falls on the same scale, you could generate them on one plot, or show them as subplots.

````python
# Each index of measures contains (homogeneity, completeness, v_measure, ARI") for a given k value
measures = np.asarray([0,0,0,0]) # create a temporary row so append works as intended
for i in range(2,11):
    preds = KMeans(n_clusters=i).fit_predict(df_data.iloc[:,:2])
    measures = np.append(measures, np.asarray(homogeneity_completeness_v_measure(df_data.target, preds) + (adjusted_rand_score(df_data.target, preds),))) # add tuples, convert to np array and append to measures
df_measures=pd.DataFrame(measures.reshape(10,4)[1:], columns=["homogeneity", "completeness", "v_measure", "ARI"])
df_measures["k"] = list(range(2,11))
df_measures
````
